```python
from scipy.stats import poisson
lam = 3
size = 10
r = poisson.rvs(mu=lam, size=10)
print(r)
```

```python
from scipy.stats import geom
p= 0.3
X = geom(p)
print(f'E[X] = %.3f' % X.mean())
print(f'Var[X] = %.3f' % X.var())
print(f'Prob(X > 5)= %.3f' % (1-X.cdf(5)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mean = 5
std = 2.5
#x-axis ranges from -3 and 3 with .001 steps
x = np.arange(mean-4*std, mean+4*std, 0.001*std)
y1 = norm.pdf(x, mean, std)
y2 = norm.cdf(x, mean, std)
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Normal density and cumulative distribution functions')
ax1.plot(x, y1)
ax1.set_title('PDF')
ax2.plot(x, y2)
ax2.set_title('CDF')
plt.show()
fig.savefig('normal.png')
```

```python
upper = norm.ppf(0.975, loc=5, scale=2.5)
lower = norm.ppf(0.025, loc=5, scale=2.5)
print(f'95% of values are between {lower} and {upper}')
```
